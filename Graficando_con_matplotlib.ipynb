```python
#pyplot es un módulo de matplotlib
#plt es un apodo de matplotlib.pyplot, entonces para invocarlo solo se escribe 'plt'

import matplotlib.pyplot as plt
```

```python
from random import randrange, seed
```

```python
list(range(1,7))
```

```python
math_grades = []
seed(4)
for grades in range(6):
  math_grades.append(randrange(1, 8))
```

```python
x = list(range(1, 7))
y = math_grades

plt.plot(x, y, marker = 'o') #para graficar con puntos
plt.title('Math Grades Graph') #título
plt.xlabel('Exams') #x axis
plt.ylabel('Grades') #y axis
plt.show() #to show only the title
```

##**Gráficos materias: matemática, lenguaje, geografía**

```python
notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_lenguaje = ['Lenguaje',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografía',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_lenguaje, notas_geografia]
```

```python
for nota in notas:
  x = list(range(1, 9))
  y = nota[1:]
  plt.plot(x, y, marker='o')
  plt.xlabel('Exámenes')
  plt.ylabel('Notas')
  plt.title(nota[0])
  plt.show()
```
